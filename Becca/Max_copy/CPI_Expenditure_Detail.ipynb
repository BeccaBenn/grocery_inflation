```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
cpi_expenditure = Path("CPI_Expenditure_detail.csv")

expenditure_df = pd.read_csv(cpi_expenditure)
```

```python
#reverse columns & rows
expenditure_df = expenditure_df.transpose()
expenditure_df
```

```python

#Change first column to df index
expenditure_df.set_index(expenditure_df.columns[0], inplace=True)

#Change first row to df columns
expenditure_df.columns = expenditure_df.iloc[0]
expenditure_df = expenditure_df.iloc[1:]
expenditure_df
```

```python
#make headers uniform to eliminate hidden spaces or symbols

expenditure_df.columns = expenditure_df.columns.str.replace(' ', '_').str.lower()
expenditure_df.columns = expenditure_df.columns.str.strip('_')
expenditure_df.columns = expenditure_df.columns.str.replace(',', '')


expenditure_df
```

```python
#drop unecessary columns

columns_to_drop = ['housing','shelter','residential_rent','fuel_and_utilities','house_furnishings_&_ops',
                  'apparel','transportation','new_vehicles','gasoline','medical_care','recreation*',
                  'education_&_communication*','other_goods_&_services',"homeowners'_equiv._rent",'cpi-u_all_items',
                  'core_cpi-u','food_&_beverages','food','food_away_from_home','alcoholic_beverages','cpi-w_all_items']

expenditure_df = expenditure_df.drop(columns=columns_to_drop, axis=1)
```

```python
expenditure_df
```

```python
#drop null columns

expenditure_df = expenditure_df.dropna(axis=1, how='all')
expenditure_df
```

```python
col_list = ['usa_food_at_home','la_food_at_home','riv_food_at_home','sd_food_at_home','sf_food_at_home']

expenditure_df.columns = col_list
```

```python
expenditure_df = expenditure_df.drop(expenditure_df.index[-4:])
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
expenditure_df
```

```python
expenditure_df = expenditure_df.fillna(0)
expenditure_df['usa_food_at_home'] = pd.to_numeric(expenditure_df['usa_food_at_home'], errors='coerce')
expenditure_df['la_food_at_home'] = pd.to_numeric(expenditure_df['la_food_at_home'], errors='coerce')
expenditure_df['riv_food_at_home'] = pd.to_numeric(expenditure_df['riv_food_at_home'], errors='coerce')
expenditure_df['sd_food_at_home'] = pd.to_numeric(expenditure_df['sd_food_at_home'], errors='coerce')
expenditure_df['sf_food_at_home'] = pd.to_numeric(expenditure_df['sf_food_at_home'], errors='coerce')
new_labels = {'usa_food_at_home':'United States','la_food_at_home':'Los Angeles','riv_food_at_home':'Riverside',
             'sd_food_at_home':'San Diego','sf_food_at_home':'San Francisco'}
for header,label in new_labels.items():
    plt.plot(expenditure_df.index, expenditure_df[header], label=label)
```

```python
# Calculate the percent change from one row to the next for all columns
percent_changes = expenditure_df.pct_change() * 100  # Multiply by 100 to convert to percentage
average_change = percent_changes[['la_food_at_home', 'riv_food_at_home', 'sd_food_at_home', 'sf_food_at_home']].mean(axis=1)
print("\nAverage Change:")
print(average_change)
# Display the DataFrame with percent changes
print("Percent Changes:")
print(percent_changes)
```

```python
plt.figure(figsize=(10, 6))

for column in percent_changes.columns:
    plt.plot(percent_changes.index, percent_changes[column], label=column)

plt.xlabel('Date')
plt.ylabel('Percent Change (%)')
plt.title('Percent Change Over Time')
plt.legend()
plt.grid(False)

plt.show()
```

```python
# Create a list of years from 2000 to 2023
columns_to_exclude = ['usa_food_at_home']
for column in percent_changes.columns:
    if column not in columns_to_exclude:
        plt.plot(percent_changes.index, percent_changes[column], label=column)
# Create a list of years from 2000 to 2023
custom_years = list(range(2000, 2024))
plt.figure(figsize=(30, 6))
for column in percent_changes.columns:
    plt.plot(percent_changes.index, percent_changes[column], label=column)
plt.xlabel('Date')
plt.ylabel('Percent Change (%)')
plt.title('Percent Change Over Time')
# Set custom x-axis ticks to display the years 2000 to 2023
plt.xticks(percent_changes.index[::12], custom_years, rotation=45)
plt.legend(['US Food Cost Percent Change','Los Angeles Food Cost Percent Change','Riverside Food Cost Percent Change',
          'San Diego Food Cost Percent Change', 'San Francisco Food Cost Percent Change'])
plt.grid(False)
plt.show()                 
```
