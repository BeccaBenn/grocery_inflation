```python
import pandas as pd
from pathlib import Path
```

```python
cpi_index = Path("historicalcpi.csv")

cpi_df = pd.read_csv(cpi_index)
```

```python
cpi_df
```

```python
new_header = cpi_df.iloc[0]
cpi_df = cpi_df[1:]
cpi_df.columns = new_header
```

```python
cpi_df = cpi_df.drop(cpi_df.index[22:26])
```

```python
cpi_df
```

```python
import matplotlib.pyplot as plt
```

```python
cpi_df.set_index('Consumer Price Index item',inplace=True)
cpi_df
```

```python
meat_df = cpi_df[3:9]
meat_df
```

```python
fruits_and_vegs_df = cpi_df[13:18]
fruits_and_vegs_df
```

```python

meat_transposed = meat_df.transpose()


meat_transposed['Meats, poultry, and fish'].plot(kind='line', marker='o', figsize=(10, 6))
plt.title('% Change in CPI by Year')
plt.xlabel('Year')
plt.ylabel('% Change')
plt.legend(title='Category')


plt.show()
```

```python
cpi_df = cpi_df.transpose()
cpi_df.columns = cpi_df.columns.str.replace(' ', '_').str.lower()
cpi_df.columns = cpi_df.columns.str.strip('_')
cpi_df.columns = cpi_df.columns.str.replace(',', '')
cpi_df
```

```python
categories = cpi_df.columns.tolist()

for cat in categories:
    print(cat)
```

```python
to_remove = ['food_away_from_home',
'other_foods',
'other_meats',
'meats_poultry_and_fish',
'nonalcoholic_beverages',
'sugar_and_sweets',
'cereals_and_bakery_products',
'fats_and_oils']

cpi_df.drop(columns=to_remove, inplace=True)
cpi_df
```

```python
graphed_categories = ['processed_fruits_and_vegetables','fresh_fruits_and_vegetables']


cpi_df[graphed_categories].plot(kind='line', marker='o', figsize=(10, 6))
plt.title('% Change in CPI by Year')
plt.xlabel('Year')
plt.ylabel('% Change')
plt.legend(title='Category')


plt.show()
```
