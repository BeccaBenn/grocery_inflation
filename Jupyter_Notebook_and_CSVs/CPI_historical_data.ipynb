```python
import pandas as pd
from pathlib import Path
```

```python
cpi_index = Path("historicalcpi.csv")

cpi_df = pd.read_csv(cpi_index)
```

```python
cpi_df
```

```python
new_header = cpi_df.iloc[0]
cpi_df = cpi_df[1:]
cpi_df.columns = new_header
```

```python
cpi_df = cpi_df.drop(cpi_df.index[22:26])
```

```python
cpi_df
```

```python
import matplotlib.pyplot as plt
```

```python
cpi_df.set_index('Consumer Price Index item',inplace=True)
cpi_df
```

```python
meat_df = cpi_df[3:9]
meat_df
```

```python
fruits_and_vegs_df = cpi_df[13:18]
fruits_and_vegs_df
```

```python
cpi_df = cpi_df.transpose()
cpi_df.columns = cpi_df.columns.str.replace(' ', '_').str.lower()
cpi_df.columns = cpi_df.columns.str.strip('_')
cpi_df.columns = cpi_df.columns.str.replace(',', '')
cpi_df
```

```python
categories = cpi_df.columns.tolist()

for cat in categories:
    print(cat)
```

```python
to_remove = ['food_away_from_home',
'other_foods',
'other_meats',
'meats_poultry_and_fish',
'nonalcoholic_beverages',
'sugar_and_sweets',
'cereals_and_bakery_products',
'fats_and_oils']

cpi_df.drop(columns=to_remove, inplace=True)
cpi_df.index = cpi_df.index.map(lambda x: str(int(x)) if x.is_integer() else str(x))
cpi_df
```

```python
# Find the 6 largest (absolute value) values in the "food_at_home" column
n_largest = 6
largest_values = cpi_df['food_at_home'].abs().nlargest(n_largest)

# Print the index and largest values
for idx in largest_values.index:
    value = largest_values[idx]
    print(f"Index: {idx}, Value: {value}")
```

```python
#  Extract the data from the DataFrame
x = cpi_df.index  # X-axis data
y = cpi_df['food_at_home']  # Y-axis data

# Create a line graph
plt.figure(figsize=(15, 10))
plt.plot(x, y, linestyle='-')
plt.xlabel('Year')
plt.ylabel('% Change')
plt.title('Largest % Change for Cost of Food at Home')
plt.grid(alpha=0.25)
plt.xticks(rotation=45)

indices_to_label = ['1975', '1978', '1979', '1980','2022']

# Label the specific data points in the 'Values' column
for index_label in indices_to_label:
    value = cpi_df.at[index_label, 'food_at_home']
    plt.text(index_label, value, f'{index_label}-{value}%', ha='right', va='bottom')
    plt.plot(index_label, value, 'ro')
    plt.savefig('C:/Users/benne/OneDrive/Desktop/BootCamp/Group_projects/project_1/grocery_inflation/Images/largest_per_change_line.png', dpi=300)

# Display the graph
plt.show()
```
